### 3.3 Empirical Rule

```python
def find_annomaly(data):
    mean = data.mean()
    std = data.std()

    upper_bound = mean + (3 * std)
    lower_bound = mean - (3 * std)

    return (data > upper_bound) | (data < lower_bound)
```

```python
nnumerical_columns = [
                        'Age', 'Tenure', 'CreditScore',
                        'Balance', 'EstimatedSalary'
                            ]

for col in nnumerical_columns:
    n_outliers = find_annomaly(df[col]).sum()
    n_outliers_percentage = round(n_outliers / len(df), 3)
    print(f"{col} : {n_outliers} outliers {n_outliers_percentage} %")
```

### 3.4 IQR Method

```python
def find_annomaly(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    return (data > upper_bound) | (data < lower_bound)
```

```python
for col in nnumerical_columns:
    n_outliers = find_annomaly(df[col]).sum()
    n_outliers_percentage = round(n_outliers / len(df), 3)
    print(f"{col} : {n_outliers} outliers {n_outliers_percentage} %")
```

Removing outlier is business decision not a technical decision. For example we cant say people above age 60 are an annomaly etc

```python
outlier_mask_log = pd.DataFrame()
for col in nnumerical_columns:
    outlier_mask_log[col] = find_annomaly(df[col])

outlier_count = outlier_mask_log.sum(axis=1)
outlier_mask_log['Outlier Count'] = outlier_count
outlier_mask_log = outlier_mask_log[outlier_mask_log['Outlier Count'] == 2]

outlier_mask_log
```
