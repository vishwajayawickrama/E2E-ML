### 2. Encode Nominal Variables

```python
nomina_variables = ['Geography', 'Gender']
geography_dummies = pd.get_dummies(df['Geography'], prefix='Geography')
gender_dummies = pd.get_dummies(df['Gender'], prefix='Gender')
df_encoded = pd.concat([df, geography_dummies, gender_dummies], axis=1)
del df_encoded['Geography']
del df_encoded['Gender']
df_encoded

```

### 3. Encode Ordinal Variables

```python
encoding_dict_creditscore = {
                                'Poor': 0,
                                'Fair': 1,
                                'Good': 2,
                                'Very Good': 3,
                                'Excellent': 4 }
# Can do easy with Scikit-learn Lable Encoder, But doint this through Programatically

df_encoded['CreditScoreBins'] = df_encoded['CreditScoreBins'].map(encoding_dict_creditscore)
df_encoded.head(5)
```

```python
df_encoded.to_csv('data/processed/ChurnModelling_Missing_Encoded.csv', index=False)
```
