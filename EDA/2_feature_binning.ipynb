Credit Score

350 - 580 => 'poor
580 - 670 => 'Fair
670 - 740 => 'Good'
740 - 800 => 'Very Good'
800+ => 'Excellent'

```python
def custom_binning_credit_score(score):
    if score < 580:
        return 'Poor'
    if score < 670:
        return 'Fair'
    if score < 740:
        return 'Good'
    if score < 800:
        return 'Very Good'
    if score <= 850:
        return 'Excellent'
    else:
        assert True, "Credit Score cant go beyind 850"

df['CreditScoreBins'] = df['CreditScore'].apply(custom_binning_credit_score)
del df['CreditScore']

df.head(5)
```

```python
df.to_csv('data/processed/ChurnModelling_Missing_Binning_Applied.csv', index=False)
```
